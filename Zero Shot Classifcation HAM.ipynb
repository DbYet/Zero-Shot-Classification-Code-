```python
# ----------------------------------------------------
# NON-GOOGLE MULTI-MODEL BENCHMARK (CORRECTED FILE)
# ----------------------------------------------------
import pandas as pd
from transformers import pipeline
from google.colab import files

print("📬 Step 1: Loading your spreadsheet...")
# CORRECTED LINE: Matching your sidebar filename perfectly down to the space!
df_results = pd.read_csv('Zero_Shot_Classifcation_All_HAM.csv')
all_emails = df_results['Clean_Message'].astype(str).tolist()

candidate_labels = ["legitimate business correspondence", "deceptive AI phishing attack"]

# ----------------------------------------------------
# RUNNING MODEL #2: DeBERTa (Meta / Microsoft / Open Source)
# ----------------------------------------------------
print("\n🧠 Step 2: Downloading Model #2 (DeBERTa-v3-Large)...")
deberta_classifier = pipeline("zero-shot-classification",
                              model="MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli",
                              device=0)

print("🚀 Running DeBERTa across your emails...")
deberta_guesses = []
deberta_scores = []

for idx, email in enumerate(all_emails):
    result = deberta_classifier(email, candidate_labels)
    deberta_guesses.append(result['labels'][0])
    deberta_scores.append(result['scores'][0])
    if idx % 100 == 0:
        print(f"🔄 DeBERTa Progress: {idx} / {len(all_emails)}")

df_results['DeBERTa_Guess'] = deberta_guesses
df_results['DeBERTa_Confidence'] = deberta_scores

# ----------------------------------------------------
# RUNNING MODEL #3: RoBERTa
# ----------------------------------------------------
print("\n🧠 Step 3: Downloading Model #3 (DistilRoBERTa)...")
# Using cross-encoder/nli-distilroberta-base as it is open and highly effective
roberta_classifier = pipeline("zero-shot-classification",
                              model="cross-encoder/nli-distilroberta-base",
                              device=0)

print("🚀 Running RoBERTa across your emails...")
roberta_guesses = []
roberta_scores = []

for idx, email in enumerate(all_emails):
    result = roberta_classifier(email, candidate_labels)
    roberta_guesses.append(result['labels'][0])
    roberta_scores.append(result['scores'][0])
    if idx % 100 == 0:
        print(f"🔄 RoBERTa Progress: {idx} / {len(all_emails)}")

df_results['RoBERTa_Guess'] = roberta_guesses
df_results['RoBERTa_Confidence'] = roberta_scores

# ----------------------------------------------------
# FINAL DISPLAY & EXPORT
# ----------------------------------------------------
print("\n==================================================")
print("🎉 MULTI-ORGANIZATIONAL STUDY COMPLETE!")
print("==================================================")

if 'AI_Guess' in df_results.columns:
    print("\n📊 Model #1 (BART) Breakdown:")
    print(df_results['AI_Guess'].value_counts())

print("\n📊 Model #2 (DeBERTa) Breakdown:")
print(df_results['DeBERTa_Guess'].value_counts())

print("\n📊 Model #3 (RoBERTa) Breakdown:")
print(df_results['RoBERTa_Guess'].value_counts())

# Save everything into one final mega file
output_filename = 'impartial_multi_model_evaluations.csv'
df_results.to_csv(output_filename, index=False)
print(f"\n💾 Saved all impartial data to '{output_filename}'")

# Trigger browser download automatically
files.download(output_filename)
```
